# 2. Extended GCD

Using the two primes p and q, find the integers u,v such that `p * u + q * v = gcd(p,q)`.

````python
p = 26513
q = 32321

# For prime numbers gcd(p, q) = 1, so p * u + q * v = 1
# Solving for u:
#    u = (1 - q * v) / p

for v in range(-p, p):
  u = (1 - q * v) / p
  if u.is_integer():
    print(u, v)
    break

# Enter whichever of u and v is the lower number as the flag.
min(u, v)
````

# 9. Chinese Remainder Theorem
Given:

```
x ≡ 2 mod 5 
x ≡ 3 mod 11 
x ≡ 5 mod 17 
```

Find the integer `a` such that `x ≡ a mod 935`.

````python
# This can only be solved with the CRT because 17*11*5=935!

a1 = 2
n1 = 5
m1 = 11*17
inv_m1 = pow(m1, -1, n1)

a2 = 3
n2 = 11
m2 = 5*17
inv_m2 = pow(m2, -1, n2)

a3 = 5
n3 = 17
m3 = 11*5
inv_m3 = pow(m3, -1, n3)

M = n1*n2*n3

solution = (a1*m1*inv_m1 + a2*m2*inv_m2 + a3*m3*inv_m3) % M
print(solution)
````
